#### if - else block

```python
age=12
if (age>=18):
    print("you are eligible for voting")
else:
    print("you are not eligible for voting")

print("program executed sucessfully")
```

#### if -elif-else block

```python
age=15

if (age<13):
    print("you are child, not eligible for voting")
elif (age<18):
    print("you are teenager, not eligible for voting")
else:
    print("you are are eligible for voting")
```
