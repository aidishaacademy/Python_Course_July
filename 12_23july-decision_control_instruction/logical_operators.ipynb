## Logical Operators

#### and operator

```python
age=22
nationality="American"

if (age>=18) and (nationality=="Indian"):
    print("you are eligible for voting")
else:
    print("you are not eligible for voting")
```

#### or operator

```python
is_weekend=False
is_holiday=False

if is_weekend or is_holiday:
    print("you can relax")
else:
    print("you need to work")
```

## conditional expression

```python
age=12
if age>=18:
    print("you are eligible for voting")
else:
    print("you are not eligible for voting")
```

```python
age=12
"you are eligible for voting" if age>=18 else "you are not eligible for voting"
```

## all ( ) - built in function

```python
a=1;b=2;c=13
```

```python
print(all((a<5,b<5,c<5)))
```

```python
age=22
nationality="indian"
voting_card=True

if all((age>=18,nationality=="indian",voting_card)):
    print("you are eligible for voting")
else:
    print("you are not eligible for voting")
```

## any ( ) - built in function

```python
age=12
nationality="indian"
voting_card=False

if any((age>=18,nationality=="indian",voting_card)):
    print("you are eligible for voting")
else:
    print("you are not eligible for voting")
```
